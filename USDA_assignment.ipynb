```python
#add a new variable/column called high sodium and make ot 1 of its
#> avergae Sodium and =0 if its<=mean(Sodium).
```

```python
df1['Sodium'].mean()
```

```python
df1['high sodium']=[1 if i>df1['Sodium'].mean() else 0 for i in df1['Sodium']] 
```

```python
df1.head()
```

```python
#similarly for high Protein , High TotalFat & high carbs
```

```python
df1['high Protein']=[1 if i>df1['Protein'].mean() else 0 for i in df1['Protein']] 
```

```python
df1['high TotalFat']=[1 if i>df1['TotalFat'].mean() else 0 for i in df1['TotalFat']] 
```

```python
df1['high Carbohydrate']=[1 if i>df1['Carbohydrate'].mean() else 0 for i in df1['Carbohydrate']] 
```
